```python
import tensorflow.compat.v1 as tf
#tf.disable_v2_behavior()
###(1) Build graph using TF operations###
#학습시킬 데이터를 입력
x_train = [1, 2, 3]
y_train = [1, 2, 3]

#w, b에게 Variable 적용 → 학습하는 과정에서 tensorflow가 스스로 변경시키는 값을 의미
#tesorflow 내부에서 자체적으로 변경시키는 값임.
W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

#our hypothsis Wx + b
hypothesis = x_train * W + b

#cost/loss function
#tf.reduce_mean(x)처럼 두번째 인자를 적지 않은 경우  변수 x가 가리키는 배열 전체  원소의 합을 원소 개수로 나누어 계산합니다.
#예를들어, t = [1., 2., 3., 4.]
#tf.reduce_mean(t) → 2.5의 값을 얻음
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

#GradientDescent (cost의 값을 minimize하자)
#Minimze
#tf.disable_eager_execution()
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

###(2) Run/update graph and get results###
#Launch the graph in a session
sess = tf.Session()

#Initializes global varibales in the graph
#tf.global_variables_initializer() : 이 부분은 W, b 사용해서 실행하기 전 반드시 실행해야 하는 부분이다.
sess.run(tf.global_variables_initializer())

for step in range(2001) : 
    sess.run(train)
    if step % 20 == 0 :
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b

print(sess.run(adder_node, feed_dict = {a : 3, b : 4.5}))
print(sess.run(adder_node, feed_dict = {a : [1, 3], b : [2, 4]}))
```

```python
#PLACEHOLDER를 사용 : 만들어지는 모델에 대해 값을 따로 넣겨줄 수 있다는 장점

W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

hypothesis = X * W + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(2001) :
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X : [1, 2, 3], Y : [1, 2, 3]})
    if step % 20 == 0 :
        print(step, cost_val, W_val, b_val)
```

```python
#testing our model
print(sess.run(hypothesis, feed_dict = {X : [5]}))
print(sess.run(hypothesis, feed_dict = {X : [2.5]}))
print(sess.run(hypothesis, feed_dict = {X : [1.5, 3.5]}))
```

```python
#다른 데이터를 넣어서 새로운 graph를 만들고 다시 테스트해보기
W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

hypothesis = X * W + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(2001) :
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X : [1, 2, 3, 4, 5], Y : [2.1, 3.1, 4.1, 5.1, 6.1]})
    if step % 20 == 0 :
        print(step, cost_val, W_val, b_val)
```

```python
#testing our model
print(sess.run(hypothesis, feed_dict = {X : [5]}))
print(sess.run(hypothesis, feed_dict = {X : [2.5]}))
print(sess.run(hypothesis, feed_dict = {X : [1.5, 3.5]}))
```
